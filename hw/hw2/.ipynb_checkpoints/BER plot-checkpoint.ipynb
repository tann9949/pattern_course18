# M-ary PSK

```python
import matplotlib.pyplot as plt
import math

nbitpersymbol = range(1,7)
EbN0dB = list(range(0,31))
EbN0 = [10**(k/10) for k in EbN0dB]
for nbps in nbitpersymbol:
    EsN0 = [k*nbps for k in EbN0]
    M = 2**nbps
    if nbps==1:
        psymbol = [0.5*math.erfc( math.sqrt(k)) for k in EsN0]
        BER = [k for k in psymbol]
    else:
        psymbol = [math.erfc( math.sqrt(k)*math.sin(math.pi/M) )
for k in EsN0]
        BER = [k/nbps for k in psymbol]
    plt.semilogy(EbN0dB,BER,label='M = '+str(M))
    print('M=',BER)
    
plt.xlabel('Eb/No (dB)')
plt.ylabel('Bit error probability (BER)')
plt.legend()
plt.ylim(10**-4,1)
plt.grid(True,'both')
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

nbitpersymbol = range(1,7)
EbN0dB = list(range(0,31))
EbN0 = [10**(k/10) for k in EbN0dB]
for nbps in nbitpersymbol:
    EsN0 = [k*nbps for k in EbN0]
    M = 2**nbps
    SER = [ (M-1)/M*math.erfc( math.sqrt(k*3*M/(M**2-1)) )
for k in EsN0]
    plt.semilogy(EbN0dB,SER,label='M = '+str(M))
    print('M=',SER)
plt.xlabel('Eb/No (dB)')
plt.ylabel('Symbol error probability (SER)')
plt.legend()
plt.ylim(10**-7,1)
plt.grid(True,'both')
plt.show()
```
